### Step 1: Import the Library
We always import pandas as `pd`. We're also importing `numpy` as `np` because pandas uses it under the hood.

```python
import pandas as pd
import numpy as np

print('Pandas successfully imported!')
```

### Step 2: Let's create some dummy mock data!
Usually, you read CSV files directly. For this tutorial, let's create a messy dataset simulating user data for an app.

```python
# Creating a dictionary of lists
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi'],
    'Age': [25, np.nan, 35, 42, 28, 55, 31, np.nan], # np.nan means missing data!
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'New York', 'Chicago', 'Miami', 'Seattle'],
    'Salary': [70000, 85000, 120000, 95000, 72000, 150000, 90000, 110000],
    'Premium_Member': [True, False, True, False, False, True, True, False]
}

# Convert it into a Pandas DataFrame
df = pd.DataFrame(data)
df
```

### Step 3: Inspecting your Data
When you load millions of rows, you can't just print `df`. You need tools to inspect it efficiently.

```python
# .head(n) shows the first n rows. Try changing it to df.head(3)
display(df.head())

# .info() tells you about missing values and data types
print('\n--- Data Info ---')
df.info()

# .describe() gives you instant math statistics on numerical columns
print('\n--- Statistics ---')
display(df.describe())
```

### Step 4: Selecting Data
How do we grab specific columns or rows?

```python
# Select a single column (Notice it returns a 'Series' - perfectly formatted 1D array)
display(df['City'])

# Select multiple columns (Pass a list of column names inside the brackets!)
display(df[['Name', 'Salary']])

display(df[["Name", "City", "Premium_Member"]])
```

### Step 5: Filtering Data 
Filtering in Pandas is incredibly powerful. We use boolean conditions inside brackets.

```python
# Let's find everyone making over 90,000
high_earners = df[df['Salary'] > 90000]
display('High Earners:', high_earners)

# Who lives in Chicago?
chicagos = df[df['City'] == 'Chicago']
display('Chicago dudes:', chicagos)

# Multiple conditions: Over 30 AND Premium Member
# Note: & is AND, | is OR. You MUST use parentheses around conditions.
target_users = df[(df['Age'] > 30) & (df['Premium_Member'] == True)]
display('Target Users:', target_users)
```

### Step 6: Handling Missing Data
Real data is messy. Notice how Bob and Heidi don't have their age listed (`NaN`). ML algorithms will crash if you feed them `NaN`.

```python
# Strategy 1: Drop any rows with missing data
df_dropped = df.dropna()
display('After Dropping Rows:', df_dropped)

# Strategy 2 (Better): Fill missing data with the average (mean) age
average_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(average_age)

display('Missing Data Filled:', df)
```

### Step 7: Feature Engineering (Creating new columns)
In ML, you often create new features to help the model learn better.

```python
# Let's calculate estimated tax (assuming 20% rate)
# Pandas does this for the ENTIRE column instantly, no for-loops needed!
df['Estimated_Tax'] = df['Salary'] * 0.20

# Let's convert City names to lowercase so our ML model maps them correctly
df['City'] = df['City'].str.lower()

df
```

### Step 8: Grouping Data (Like SQL GroupBy)
Grouping allows you to calculate statistics across categories.

```python
# What is the average salary per city?
avg_salary_by_city = df.groupby('City')['Salary'].mean()
display('Average Salary by City:', avg_salary_by_city)

# Are premium members older on average?
age_by_membership = df.groupby('Premium_Member')['Age'].mean()
display('Average Age by Membership:', age_by_membership)
```

## Complex Query
I need a table showing only the 'Electronics' and 'Home' categories. I want to see the total number of items sold, the average price of those items, and the number of returned items per category. Oh, and only include purchases over $50. Finally, please sort the results so the category with the highest average price is at the top.

```python
np.random.seed(42) # So we get the same random numbers
sales_data = {
    'Customer_ID': np.random.randint(1000, 1020, 100),
    'Category': np.random.choice(['Electronics', 'Clothing', 'Home', 'Toys'], 100),
    'Price': np.random.uniform(10, 500, 100).round(2),
    'Is_Returned': np.random.choice([True, False], 100, p=[0.15, 0.85]) # 15% return rate
}
sales_df = pd.DataFrame(sales_data)

sales_df
```

```python
report = (
    # Notice we combine them inside ONE pair of brackets using '&'
    sales_df[sales_df["Price"] > 50] [sales_df["Category"].isin(["Electronics", "Home"])]
    .groupby("Category")
    .agg(
        Total_Items_Sold=('Customer_ID', 'count'), # Count total rows per category
        Average_Price=('Price', 'mean'),           # Find average price
        Items_Returned=('Is_Returned', 'sum')      # Summing booleans counts the True values!
    )
    .sort_values(by="Average_Price", ascending=False)
    .round(2)
    )

display(report)
```
