Pandas provides us with using its built in indexing methods to access specific rows and columns, specifically the `loc` and `iloc`. <br> <br>
`loc` stands for **Location** and it understands the labels or names of the row/column. <br> <br>
`iloc` stands for **Integer Location** and it only understands the raw number positions, like standard Python lists.

#### The Problem
Imagine you have this simple DataFrame of 3 houses. Notice the row labels (the index) are custom string names. How do we get the price of House_C?

iloc only takes the row and column number as input.

Syntax:
`df.iloc[row_number, column_number]`

```python
df.iloc[2,1]
```

```python
# Get the price of all houses
df.iloc[:,1]
```

```python
df.iloc[:1, 1]
```

```python
# get the first row in the dataframe
df.iloc[0]
```

Both `loc` and `iloc` are row-first, column-second.

```python
# Get the third and fourth entries
df.iloc[2:4]
```

```python
# Can also go backwards, getting the last 5 houses
df.iloc[-5:]
```

```python
# Get price of first listing
df.loc[0,'listPrice']
```

### The "Gotcha"
If you don't assign custom names to your rows (which most people don't), your row labels default to 0, 1, 2, 3...

Because of this, .loc[0] and .iloc[0] look like they are doing the same thing. But they aren't!

.iloc[0] means "Give me the first row in the file".
.loc[0] means "Give me the row whose custom name plate happens to be '0'".
If you sort or shuffle your data, the row named '0' might now be at the bottom of the table. In that case, .iloc[0] grabs the new top row, but .loc[0] shoots down to the bottom to find the row with the name plate '0'.

If we do .loc(0:100), it will actually return 101 rows, both indexes are inclusive, since the loc function returns the rows named '0' all the way to the row named '100'
