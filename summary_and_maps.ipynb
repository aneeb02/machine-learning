# 🗺️ Summary Functions and Maps in Pandas

Once you have isolated your data, you often need to either **summarize** it (get the big picture) or **map** it (transform every single value).

```python
import pandas as pd
import numpy as np

np.random.seed(0)
data = {
    'country': ['Italy', 'Portugal', 'US', 'Spain', 'Italy', 'France', 'US', 'Spain', 'France', 'Italy'],
    'points': [87, 90, 85, 95, 92, 88, 93, np.nan, 84, 91],
    'price': [15, 22, 10, 45, 30, np.nan, 25, 55, 12, 35],
    'variety': ['White', 'Red', 'White', 'Red', 'Red', 'White', 'Red', 'Red', 'White', 'Red']
}
reviews = pd.DataFrame(data)
reviews
```

## 1. Summary Functions
These functions take many values and boil them down to a single summary value.

```python
# describe() is the most common. It gives statistical summaries.
display("Describe Numerical:", reviews.describe())

# You can also use describe on string/categorical columns!
display("Describe Strings:", reviews['country'].describe())
```

```python
# Individual summary metrics:
mean_price = reviews['price'].mean()
display(f"Mean Price: {mean_price}")

unique_countries = reviews['country'].unique()
display("Unique Countries:", unique_countries)

# value_counts() is incredibly useful for seeing distributions!
country_counts = reviews['country'].value_counts()
display("Country Value Counts:", country_counts)
```

## 2. Maps (`map` and `apply`)
A "map" is a mathematical term for a function that takes one set of values and "maps" them to another set. In data science we often need to create new representations from existing data, or transform data from the format it is in now to the format that we want it to be in later.

### `map()`
`map()` is usually used on a single **Series** (a single column).

```python
# Let's center the prices around the mean.
review_price_mean = reviews['price'].mean()

centered_price = reviews['price'].map(lambda p: p - review_price_mean)
display("Centered Prices:", centered_price)
```

### `apply()`
`apply()` is the equivalent method if we want to transform a whole DataFrame by calling a custom method on each row.

```python
def remean_points(row):
    row['points'] = row['points'] - reviews['points'].mean()
    return row

# axis='columns' means we apply the function across each row
reviews_remeaned = reviews.apply(remean_points, axis='columns')
display("DataFrame with Remeaned Points:", reviews_remeaned)
```

### Faster Mapping (Vectorization)
Pandas provides many built-in mapping operations that are often faster than `map()` or `apply()`.

```python
# This is equivalent to our map() example above, but MUCH faster.
review_price_mean = reviews['price'].mean()
fast_centered_price = reviews['price'] - review_price_mean

# We can even combine columns entirely!
country_and_variety = reviews['country'] + " - " + reviews['variety']
display("Combined Strings:", country_and_variety)
```
