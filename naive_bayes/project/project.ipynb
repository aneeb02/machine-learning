```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score
```

Scikit-learn's GaussianNB requires a flattened array, not a DF column

```python
y_flat = y['Class'].values  # converts the DF column into a 1-D array
y_flat
```

Splitting our data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_flat, test_size=0.2, random_state=42)
```

Creating and initializing our model

```python
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
print(f'Classification Report: \n {classification_report(y_test, predictions, target_names=["Not Spam", "spam"])}')
```

```python
print(f'Accuracy Score: \n {accuracy_score(y_test, predictions):.2%}')
```

```python
print(f'Precision Score: \n {precision_score(y_test, predictions):.2%}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
cm = confusion_matrix(y_test, predictions)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Predicted Not Spam', 'Predicted Spam'],
            yticklabels=['Actual Not Spam', 'Actual Spam'])
plt.title('Naive Bayes Confusion Matrix')
plt.show()
```
